**Bonus Task:** Use Slicer Python to analyze the image data from Task 3 (ultrasound.nrrd). [33 Points]

```
# Activate the console with View -> Python Interactor.
# Now, you can run Python commands in 3D Slicer (don't run the commands here but in 3D Slicer!).
#

#
# 1) Grab the image data as a numpy array.
# Hint: You need to replace $MRML_ID with the correct value.
# Hint: You can find that value in the DATA module, if you activate 'Show MRML IDs'
node = slicer.mrmlScene.GetNodeByID('$MRML_ID')
data_as_array = slicer.util.arrayFromVolume(node)

#
# 2) Now, data_as_array is a numpy array containing the image data.
#    We can print the shape and also the max value.
print(data_as_array.shape)
# ANS: (139, 512, 512)
print(data_as_array.max())
# ANS: 3071

#
# 3) Please write numpy code to find the coordinates of the single pixel
# with the maximum value.
import numpy as np
max_index = np.unravel_index(data_as_array.argmax(), data_as_array.shape)
print("Coordinates of pixel with maximum value:", max_index)

# ANS: Coordinates of pixel with maximum value: (110, 146, 403)

```
